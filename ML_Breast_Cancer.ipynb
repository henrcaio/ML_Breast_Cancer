```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("breast_cancer.csv")
data.head()
```

```python
data.info()
```

```python
data.isna().sum().sum()
```

```python
data.describe().T
```

```python
data.drop(columns='id', inplace=True)
```

```python
data.columns
```

```python
X = data.drop(columns='diagnosis')
y = data['diagnosis']
```

```python
X.corr()
```

```python
le = LabelEncoder()
y_encoded=le.fit_transform(y)
```

```python
reg_log = LogisticRegression()

xtr, xts, ytr, yts = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

reg_log.fit(xtr, ytr)

y_pred = reg_log.predict(xts)
```

```python
# Usando EN, KFold e CV

n = 5
en = ElasticNet()

kf = KFold(n_splits=n, shuffle=True, random_state=42)

result_cv = cross_validate(
  estimator=en, X=X, y=y_encoded, scoring='roc_auc', cv=kf, return_train_score=True
)

df_result_cv = pd.DataFrame(result_cv)
df_result_cv
```

```python
df_result_cv.describe().T
```

```python
display(df_result_cv[["train_score", "test_score"]])

plt.title("\n DistribuiÃ§Ã£o de mÃ©tricas de treino")
sns.histplot(data=df_result_cv, x="train_score", kde=True)
plt.show()

plt.title("\n DistribuiÃ§Ã£o de mÃ©tricas de teste")
sns.histplot(data=df_result_cv, x="test_score", kde=True)
plt.show()


sns.histplot(data=df_result_cv, x="train_score", kde=True)
sns.histplot(data=df_result_cv, x="test_score", kde=True, color="orange")
plt.show()
```
